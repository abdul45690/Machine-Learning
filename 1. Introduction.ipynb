# Q2- What is supervised learning? List some examples of supervised learning.

Learns to make predictions or decisions based on labeled training data. In supervised learning, the algorithm is trained using input-output pairs.The output can be classification or regression(continous value) type. EX : 

- Spam Detection: Given a set of labeled emails, the goal is to train a model that can classify new, unseen emails as either spam or not spam.

- Fraud Detection: Given a set of labeled credit card transactions, the goal is to train a model that can detect fraudulent transactions in real-time.

- Sentiment Analysis: Given a set of labeled reviews, the goal is to train a model that can classify new, unseen reviews as either positive, negative, or neutral.

- Medical Diagnosis: Given a set of labeled medical records, the goal is to train a model that can predict whether a patient has a particular disease based on their symptoms and medical history.



# Q3- What is unsupervised learning? List some examples of unsupervised learning.
Learns from unlabelled dataset . The goal of unsupervised learning is to identify patterns, relationships, or structures within the data to gain insights or make predictions. EX: 
- Clustering: Clustering algorithms group similar data points together based on their features or attributes.
- Dimensionality Reduction: Dimensionality reduction algorithms reduce the number of features or variables in a dataset while preserving the most important information. This can help simplify data visualization and modeling. For example, principal component analysis (PCA) is a commonly used dimensionality reduction technique.
- Anomaly detection: Anomaly detection algorithms identify unusual or rare data points that deviate from the norm. For example, banks might use anomaly detection to identify fraudulent transactions.
- Association rule mining: Association rule mining algorithms identify patterns or relationships between different items or features in a dataset
- Generative models: Generative models learn to generate new data that is similar to the input data

# Q7- How can unsupervised learning be used in anomaly detection?
Unsupervised learning can be used in anomaly detection by using clustering algorithms to identify unusual patterns or outliers in the data. Anomalies are often rare events that differ significantly from the majority of the data. They can be caused by errors, malfunctions, or unusual behaviors that are not captured by the regular patterns in the data.

Here are some common approaches to using unsupervised learning in anomaly detection:
- `Clustering`: Clustering algorithms group similar data points together based on their features. Anomalies are then identified as data points that do not belong to any of the clusters or belong to a cluster with a significantly different distribution. This approach is suitable for detecting anomalies in high-dimensional data where it is difficult to define a normal pattern.
- `Density-based methods`: Density-based methods identify anomalies as data points that have a low probability of belonging to the underlying data distribution. This approach is suitable for detecting anomalies in spatial or temporal data.
- `Autoencoders`: Autoencoders are neural network models that can learn to reconstruct the input data from a compressed representation. Anomalies are then identified as data points that have a high reconstruction error, indicating that they deviate significantly from the regular patterns in the data. This approach is suitable for detecting anomalies in high-dimensional data with complex patterns.
